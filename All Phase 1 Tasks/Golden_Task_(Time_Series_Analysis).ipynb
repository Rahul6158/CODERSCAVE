# Time_Series_Analysis

[**Dataset**](https://www.kaggle.com/datasets/hershyandrew/amzn-dpz-btc-ntfx-adjusted-may-2013may2019)

```python
import pandas as pd
import matplotlib
from matplotlib import pyplot
```

```python
data=pd.read_csv('portfolio_data.csv')
print(data.head())
```

```python
print(data.notnull().sum())
```

```python
print(data.isnull().sum())
```

```python
data['Date'] = pd.to_datetime(data['Date'] + ' 00:00:00')
data.set_index('Date',inplace=True)
resamp_data = data.resample('D').mean()
```

```python
print(resamp_data)
```

```python
resamp_data.plot(figsize=(10,4))
plt.title("Time Analysis Date")
plt.xlabel("Data")
plt.ylabel("Value")
plt.legend(resamp_data.columns)
plt.show
```
